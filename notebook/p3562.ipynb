```python
import networkx as nx
import matplotlib.pyplot as plt

# LeetCode 常见输入
edges = [[1, 3], [1, 24], [3, 22], [1, 14], [3, 26], [22, 21], [14, 12], [26, 11], [21, 18], [26, 8], [26, 27], [26, 6],
         [27, 23], [8, 13], [13, 19], [21, 10], [24, 2], [13, 4], [6, 17], [23, 29], [12, 25], [11, 20], [27, 5],
         [20, 7], [10, 15], [17, 9], [10, 16], [15, 28]]

G = nx.Graph()  # 无向图
G.add_edges_from(edges)

plt.figure(figsize=(5, 5))
nx.draw(
    G,
    with_labels=True,
    node_size=1500,
    node_color="lightblue",
    font_size=12
)
plt.show()
```
